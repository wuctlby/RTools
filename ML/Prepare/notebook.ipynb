```python
import os
import subprocess
for subpath in ['Data4MC3', 'Data4MC3_nonorm', 'Data4MC3_nopid_nonorm']:
# subpath = 'Data4MC3'
    path = f'/home/wuct/ALICE/local/RTools/RTools/ML/Trainning/config/{subpath}'
    configs = [f for f in os.listdir(path) if f.endswith('.yml')]

    for config in configs:
        config_path = os.path.join(path, config)
        if not os.path.exists(config_path):
            print(f"Config file not found: {config_path}")
            continue
        
        python_command = [
            'python3',
            '/home/wuct/ALICE/local/RTools/RTools/ML/Trainning/MLClassfication_run3.py',
            config_path,
            '--train'
        ]
        
        print(f'Executing: {" ".join(python_command)}')
        
        try:
            result = subprocess.run(python_command, check=True, capture_output=True, text=True)
            print("Output:", result.stdout)
        except subprocess.CalledProcessError as e:
            print(f"Error executing command: {python_command}")
            print("Return code:", e.returncode)
            print("Error output:", e.stderr)
```

```python
import pickle
import matplotlib.pyplot as plt
import numpy as np

def plot_roc_curves_final(average_roc):
    import matplotlib.pyplot as plt
    plt.figure(figsize=(12, 7))
    all_edges = set()
    all_auc_values = []
    plot_data = {}

    for subpath, data_list in average_roc.items():
        data_list = sorted(data_list, key=lambda x: x['pt_range'][0])
        x = [item['pt_range'][0] for item in data_list]
        x.append(data_list[-1]['pt_range'][1])
        y = [item['roc_auc'] for item in data_list]
        y.append(y[-1])

        for item in data_list:
            all_edges.update(item['pt_range'])

        line = plt.step(
            x, y, where='post', label=subpath, linewidth=2.5, alpha=0.8
        )
        plot_data[subpath] = {
            'line': line[0],
            'auc_values': y,
            'bin_centers': [sum(item['pt_range'])/2 for item in data_list]
        }

        for item in data_list:
            all_auc_values.append((item['roc_auc'], subpath, item['pt_range']))


    min_edge = min(all_edges)
    max_edge = max(all_edges)
    bin_extension = (max_edge - min_edge) * 0.1
    plt.xlim(min_edge - bin_extension, max_edge + bin_extension)

    aucs = [auc for auc, _, _ in all_auc_values]
    plt.ylim(max(0.8, min(aucs) - 0.02), min(1.0, max(aucs) + 0.12))
    # plt.xlim(0.5, 26)

    all_auc_values.sort(key=lambda x: (x[2][0], x[0]))
    vertical_offset = 0
    bin_offset_dict = {}
    for auc, subpath, pt_range in all_auc_values:
        bin_center = sum(pt_range) / 2
        
        if bin_center not in bin_offset_dict:
            bin_offset_dict[bin_center] = 0
        else:
            bin_offset_dict[bin_center] += 1

        vertical_offset = bin_offset_dict[bin_center] * 0.008
        
        color = plot_data[subpath]['line'].get_color()
        plt.text(
            x=bin_center,
            y=auc + 0.01 + vertical_offset,
            s=f"{auc:.3f}",
            color=color,
            ha='center',
            fontsize=10,
            fontweight='bold',
            bbox=dict(
                facecolor='white',
                alpha=0.8,
                edgecolor=color,
                boxstyle='round,pad=0.3'
            )
        )

    plt.xlabel('pT [GeV/c]', fontsize=12)
    # plt.xscale('log')
    plt.ylabel('ROC AUC', fontsize=12)
    # plt.title('Model Performance Across pT Bins', fontsize=14, pad=15)
    plt.grid(True, axis='y', linestyle=':', alpha=0.4)
    plt.legend(
        # title='Training Config',
        loc='lower right',
        frameon=True,
        edgecolor='black'
    )
    # 添加bin分割线
    for edge in sorted(all_edges):
        plt.axvline(edge, color='gray', linestyle=':', alpha=0.3, linewidth=1)
    plt.tight_layout()
    plt.show()

path = '/home/wuct/ALICE/local/RTools/RTools/ML/Trainning/results'
subpaths = [sub for sub in os.listdir(path) if os.path.isdir(os.path.join(path, sub)) and sub.startswith('Data4MC3')]
average_roc = {}
for subpath in subpaths:
    for root, dirs, files in os.walk(os.path.join(path, subpath)):
        for file in files:
            if file.endswith('.pkl') and file.startswith('ROCCurveAll_pT_'):
                with open(os.path.join(root, file), 'rb') as f:
                    fig = pickle.load(f)
                import matplotlib.pyplot as plt
                plt.close(fig)
                pt_range = file.replace('ROCCurveAll_pT_', '').replace('.pkl', '').split('_')
                axes = fig.get_axes()
                ax = axes[0]
                legend = ax.get_legend()
                labels = [text.get_text() for text in legend.get_texts()]
                for label in labels:
                    if 'Average OvO ROC AUC' in label:
                        roc_auc = float(label.split(':')[-1].strip())
                        if subpath not in average_roc:
                            average_roc[subpath] = []
                        average_roc[subpath].append({
                            'pt_range': [float(pt_range[0]), float(pt_range[1])],
                            'roc_auc': roc_auc
                        })
print(average_roc)
plot_roc_curves_final(average_roc)
```
